```python
import numpy as np
import matplotlib.pyplot as plt

def objective(X):
    #x = trial.suggest_uniform("x", 0, 5) # 変数xを上下限0~5の範囲で連続値
    #y = trial.suggest_uniform("y", 0, 3) # 変数yを上下限0~3の範囲で連続値
    x, y = X[:,0], X[:,1]
    v0 = 4 * x ** 2 + 4 * y ** 2
    v1 = (x - 5) ** 2 + (y - 5) ** 2
    return np.c_[-v0, -v1]
```

```python
x = np.linspace(0, 5, 100)
y = np.linspace(0, 3, 100)

X, Y = np.meshgrid(x, y)
```

# 最適化の実行

```python
import numpy as np
import matplotlib.pyplot as plt
import physbo
import itertools
```

## 探査空間の定義
ここがわからん。定義する必要ある？入力変数多くなったらここで発散しそうだけど、、、

```python
x = np.linspace(0, 5, 10)
y = np.linspace(0, 3, 10)
test_X = np.array(list(itertools.product(x, y)))
```

### シミュレータの定義

```python
class simulator(object):
    def __init__(self, X):
        self.t = objective(X)

    def __call__( self, action):
        return self.t[action]

sim = simulator(test_X)
#sim(2)とかでとりだせる
```

```python
policy = physbo.search.discrete_multi.policy(test_X=test_X, num_objectives=2)
policy.set_seed(0)

policy.random_search(max_num_probes=5, simulator=sim)
res_HVPI = policy.bayes_search(max_num_probes=200, simulator=sim, interval=10, num_rand_basis=500)
# intervalって多分ハイパーパラメータチューニングの間隔かな
```

```python
front, front_num = res_HVPI.export_pareto_front()
front, front_num
```

```python
def plot_pareto_front(res):
    front, front_num = res.export_pareto_front()
    dominated = [i for i in range(res.num_runs) if i not in front_num]
    points = res.fx[dominated, :]

    plt.figure(figsize=(7, 7))
    plt.scatter(res.fx[dominated,0], res.fx[dominated,1], c = "blue")
    plt.scatter(front[:, 0], front[:, 1], c = "red")
    plt.title('Pareto front')
    plt.xlabel('Objective 1')
    plt.ylabel('Objective 2')
    #plt.xlim([-1.0, 0.0])
    #plt.ylim([-1.0, 0.0])

plot_pareto_front(res_HVPI)
```
