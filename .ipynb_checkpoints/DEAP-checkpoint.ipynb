# 関数の定義

```python
def beam(X):#(T, B, L, H):#, F, S):
    T, B, L, H = X[0], X[1], X[2], X[3]#, X[4], X[5]
    Wc, Mc = 1, 1.5
    Young = 200 #(E/GPa) 鉄のヤング率参照
    F = 10
    S = 500
    Wvol = H ** 2 * L
    Mvol = T * B * (S + L)
    Cost = (Wc * Wvol + Mc * Mvol) #最小化
    Area = L * H * np.sqrt(2)
    t1 = (F * (S + L/2))/(B * Area)
    t2 = F/(2 * Area)
    Disp = (F * S ** 3)/(4 * T ** 3 * B * Young) #最小化 
    MaxS = np.sqrt(t1 ** 2 + t2 ** 2)
    MaxT = (6 * S *F)/(T ** 2 * B)
    
    return Disp, Cost#, MaxS, MaxT, 
```

```python
import array
import random
import json

import numpy as np
import matplotlib.pyplot as plt

from math import sqrt

from deap import algorithms
from deap import base
from deap import benchmarks
from deap.benchmarks.tools import diversity, convergence, hypervolume
from deap import creator
from deap import tools

creator.create("FitnessMin", base.Fitness, weights=(-1.0, -1.0))
creator.create("Individual", array.array, typecode='d', fitness=creator.FitnessMin)

toolbox = base.Toolbox()
#BOUND_LOW, BOUND_UP = 0.0, 100.0
BOUND_LOW = [0.1, 0.1, 0.1, 0.1]
BOUND_UP = [100., 100., 30., 20.]

NDIM = 4

def uniform(low, up, size=None):
    try:
        return [random.uniform(a, b) for a, b in zip(low, up)] #リストで与えられたときはこっち
    except TypeError:
        return [random.uniform(a, b) for a, b in zip([low] * size, [up] * size)]#スカラーのときはこっち

toolbox.register("attr_float", uniform, BOUND_LOW, BOUND_UP, NDIM)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.attr_float)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", beam)#benchmarks.zdt1)
toolbox.register("mate", tools.cxSimulatedBinaryBounded, low=BOUND_LOW, up=BOUND_UP, eta=20.0)
#toolbox.register("mate", tools.cxBlend, alpha=0.2)
#toolbox.register("mutate", tools.mutGaussian, mu=[0.0, 0.0, 0.0, 0.0], sigma=[10.0, 10.0, 10.0, 10.0], indpb=0.2)
toolbox.register("mutate", tools.mutPolynomialBounded, low=BOUND_LOW, up=BOUND_UP, eta=20.0, indpb=1.0/NDIM)
toolbox.register("select", tools.selNSGA2)

def main(seed=None):
    random.seed(seed)

    NGEN = 250
    MU = 100
    CXPB = 0.9

    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("min", np.min, axis=0)
    stats.register("max", np.max, axis=0)
    
    logbook = tools.Logbook()
    logbook.header = "gen", "evals", "min", "max"
    
    #個体集団の生成
    pop = toolbox.population(n=MU)
    pop_ini = pop[:] #初期個体はpop_iniに別で保存
    
    invalid_ind = [ind for ind in pop if not ind.fitness.valid]#適応度(=関数)適用していない個体の抽出
    fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit
        
    pop = toolbox.select(pop, len(pop))#次世代個体の選択・複製
    
    record = stats.compile(pop)
    logbook.record(gen=0, evals=len(invalid_ind), **record)
    print(logbook.stream)

    for gen in range(1, NGEN):

        offspring = tools.selTournamentDCD(pop, len(pop))
        offspring = [toolbox.clone(ind) for ind in offspring]
        
        for ind1, ind2 in zip(offspring[::2], offspring[1::2]):#子供の偶数indexと奇数indexを交叉
            if random.random() <= CXPB:
                toolbox.mate(ind1, ind2)
            
            toolbox.mutate(ind1) #今の場合、交差したら必ず変異している？
            toolbox.mutate(ind2)
            del ind1.fitness.values, ind2.fitness.values #交差させた個体は適応度を削除
        
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid] #削除したので再度適応度を計算
        fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
        for ind, fit in zip(invalid_ind, fitnesses):
            ind.fitness.values = fit

        pop = toolbox.select(pop + offspring, MU)#個体集団を更新(pop+offの中からMU個だけ選択する)
        record = stats.compile(pop)
        logbook.record(gen=gen, evals=len(invalid_ind), **record)
        if gen % 10 == 0: #あんまり関係なかった
            print(logbook.stream)
            #print(gen)

    print("Final population hypervolume is %f" % hypervolume(pop))#, [0.0, 0.0, 0.0, 0.0]))

    return pop, pop_ini, logbook
        
if __name__ == "__main__":
    pop, pop_ini, stats = main(64)

    fitnesses_ini = np.array([list(pop_ini[i].fitness.values) for i in range(len(pop_ini))])
    fitnesses = np.array([list(pop[i].fitness.values) for i in range(len(pop))])

    plt.plot(fitnesses_ini[:,0], fitnesses_ini[:,1], "b.", label="Initial")
    plt.plot(fitnesses[:,0], fitnesses[:,1], "r.", label="Optimized" )
    plt.legend(loc="upper right")
    plt.title("fitnesses")
    plt.xlabel("f1")
    plt.ylabel("f2")
    plt.grid(True)
    #plt.savefig("fitnesses.png", dpi=300)
```

```python
def f(X):
    x, y = X[0], X[1]
    Z = 0.1*(x**2+y**2-16)**2 + 10*np.sin(3*x)
    return Z, #返り値が1つのときにはタプルにするためにカンマを入れる
```

```python
import array, random, json
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from deap import algorithms
from deap import base
from deap import benchmarks
from deap.benchmarks.tools import diversity, convergence, hypervolume
from deap import creator
from deap import tools

creator.create("FitnessMin", base.Fitness, weights=(1.0, ))#目的関数の返り値の数だけweightを設定する。1.:最大値, -1.:最小値
creator.create("Individual", array.array, typecode='d', fitness=creator.FitnessMin)

toolbox = base.Toolbox()
#BOUND_LOW, BOUND_UP = 0.0, 100.0
BOUND_LOW = [-3., -3.]
BOUND_UP = [3., 3.]

NDIM = 2 #最適化する入力変数

def uniform(low, up, size=None):
    try:
        return [random.uniform(a, b) for a, b in zip(low, up)] #リストで与えられたときはこっち
    except TypeError:
        return [random.uniform(a, b) for a, b in zip([low] * size, [up] * size)]#スカラーのときはこっち

toolbox.register("attr_float", uniform, BOUND_LOW, BOUND_UP, NDIM)
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.attr_float)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", f)#benchmarks.zdt1)
toolbox.register("mate", tools.cxSimulatedBinaryBounded, low=BOUND_LOW, up=BOUND_UP, eta=1.0)
#toolbox.register("mate", tools.cxBlend, alpha=0.2)
#toolbox.register("mutate", tools.mutGaussian, mu=[0.0, 0.0], sigma=[2.0, 2.0], indpb=0.2)
toolbox.register("mutate", tools.mutPolynomialBounded, low=BOUND_LOW, up=BOUND_UP, eta=1.0, indpb=1.0/NDIM)
toolbox.register("select", tools.selNSGA2)

def main(seed=None):
    random.seed(seed)

    NGEN = 250
    MU = 100
    CXPB = 0.4
    MUT_PB = 0.2
    
    #記録の設定
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("min", np.min, axis=0)
    stats.register("max", np.max, axis=0)
    
    logbook = tools.Logbook()
    logbook.header = "gen", "evals", "min", "max"
    
    #個体集団の生成
    pop = toolbox.population(n=MU)
    pop_ini = pop[:] #初期個体はpop_iniに別で保存
    
    invalid_ind = [ind for ind in pop if not ind.fitness.valid]#適応度(=関数)適用していない個体の抽出
    fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit
        
    pop = toolbox.select(pop, len(pop))#次世代個体の選択・複製
    
    record = stats.compile(pop)
    logbook.record(gen=0, evals=len(invalid_ind), **record)
    print(logbook.stream)

    result = [] #目的関数格納用
    for gen in range(1, NGEN):

        offspring = tools.selTournamentDCD(pop, len(pop))
        offspring = [toolbox.clone(ind) for ind in offspring]
        
        for ind1, ind2 in zip(offspring[::2], offspring[1::2]):#子供の偶数indexと奇数indexを交叉
            if random.random() <= CXPB:
                toolbox.mate(ind1, ind2)
            
#             toolbox.mutate(ind1) #今の場合、交差したら必ず変異している？
#             toolbox.mutate(ind2)
#             del ind1.fitness.values, ind2.fitness.values #交差させた個体は適応度を削除
            
        # 変異
        for mutant in offspring:
            # 変異させる個体を選択
            if random.random() < MUT_PB:
                toolbox.mutate(mutant)

                # 変異させた個体は適応度を削除する
                del mutant.fitness.values
        
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid] #削除したので再度適応度を計算
        fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
        for ind, fit in zip(invalid_ind, fitnesses):
            ind.fitness.values = fit

        result.append(pop[0].fitness.values)#個体0の値を格納
        pop = toolbox.select(pop + offspring, MU)#個体集団を更新(pop+offの中からMU個だけ選択する)
        print(gen, pop[0])
        record = stats.compile(pop)
        logbook.record(gen=gen, evals=len(invalid_ind), **record)
        if gen % 10 == 0: #あんまり関係なかった
            #print(logbook.stream)
            print(gen)

    print("Final population hypervolume is %f" % hypervolume(pop))#, [0.0, 0.0, 0.0, 0.0]))

    return pop, pop_ini, logbook, result

#実行
if __name__ == "__main__":
    pop, pop_ini, stats, result = main(123)

    fitnesses_ini = np.array([list(pop_ini[i].fitness.values) for i in range(len(pop_ini))])
    fitnesses = np.array([list(pop[i].fitness.values) for i in range(len(pop))])

    ##############後処理################
    #関数の定義
    x = np.linspace(-3, 3, 100)
    y = np.linspace(-3, 3, 100)

    X, Y = np.meshgrid(x, y)
    Z = 0.1*(X**2+Y**2-16)**2 + 10*np.sin(3*X)#Zは山の高さ

    #グラフのプロット
    plt.figure(figsize=(7, 7))
    cont = plt.contour(X, Y, Z, cmap="Greys")
    cont.clabel(fmt='%1.1f', fontsize=14)
    plt.xlabel("X")
    plt.ylabel("Y")

    for i in range(len(pop)):
        plt.scatter(pop[i][0], pop[i][1], color="red")
        plt.scatter(pop_ini[i][0], pop_ini[i][1], color="blue")

    plt.show()
```
